```python
from bs4 import BeautifulSoup
import requests as req
import pandas as pd

url = 'https://www.yellowpages.com/chicago-il/mip/cafe-il-cortile-8731718/'
resp = req.get(url)
data = resp.text
soup = BeautifulSoup(data,'html.parser')
```

```python
adres = []
rate = []
infos = soup.find_all(class_='primary-info')
for info in infos:
    address = info.find('h2',class_='address').text
    adres.append(address)
    rating = info.find(class_='ratings').text
    rate.append(rating)
```

```python
site = []
times = []
manu = []
linkings = soup.find_all(class_='business-card-footer')
for link in linkings:
    website = link.find('a',class_='primary-btn').get('href')
    site.append(website)
    web_r = req.get(website)
    web_d = web_r.text
    web_s = BeautifulSoup(web_d,'html.parser')
    time = web_s.find_all(class_='p af')
    for i in time: 
        i.text
        times.append(i.text.replace(',',''))
    #menu = link.find(class_='view-menu').get('href')
    #men_r = req.get(menu)
    #men_d = men_r.text
    #men_s = BeautifulSoup(men_d,'html.parser')
    #menus = men_s.find_all(class_='title-price')
    #for i in menus:
        #i.text.replace(u'\xa0', u' ')
        #manu.append(i.text.replace(u'\xa0', u' '))
```

```python
al = []

info = soup.find_all('dl')

for re in info[0].find_all('dt'):
    al.append(re.text.replace(u'\xa0', u' '))
bl = []
for i in info[0].find_all('dd'):
    bl.append(i.text.replace(u'\xa0', u' '))
al_bl = []
for x,y in zip(al,bl):
    al_bl.append({'info tittle':x,'detail info':y})
```

```python
ad = pd.DataFrame({'Address':adres})
ra = pd.DataFrame({'rating':rate})
si = pd.DataFrame({'Website':site})
ti = pd.DataFrame({'Pick hour':times})
abl = pd.DataFrame(al_bl)
#ma = pd.DataFrame({'Menu':manu})
df = pd.concat([ad,ra,si,ti,abl],axis = 1)
df.to_csv('cortile.csv',encoding = 'utf-8',header=True)
df
```
